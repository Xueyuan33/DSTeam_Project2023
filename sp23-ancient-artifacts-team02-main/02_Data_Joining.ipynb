# 02_Data_Joining, Team 02

## 1—Import Libraries

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None 
```

# 2—Read in Preprocessed Datasets

```python
lithic_p_df = pd.read_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/lithic_preprocessed_final.csv')
lithic_p_df.head(5)
```

```python
# unsure why there is an extra first column, let's drop that
lithic_p_df = lithic_p_df.drop(lithic_p_df.columns[0], axis=1)
lithic_p_df.head(5)
```

```python
soil_p_df = pd.read_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/all_soil_preprocessed.csv')
soil_p_df.head(5)
```

```python
# unsure why there is an extra first column, let's drop that
soil_p_df = soil_p_df.drop(soil_p_df.columns[0], axis=1)
soil_p_df.head(5)
```

# 3—Create New 'Lithic' Variable for each Dataset
## 1 = lithic particle, 0 = soil particle 

```python
lithic_p_df = lithic_p_df.assign(lithic = 1)
soil_p_df = soil_p_df.assign(lithic = 0)
```

```python
lithic_p_df.head(5)
```

```python
soil_p_df.head(5)
```

# 4—Join the Datasets

```python
joined_df = pd.concat([lithic_p_df, soil_p_df], ignore_index=True)
joined_df
```

# 5—Save the Joined Dataset

```python
joined_df.to_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/lithic_all_soil_joined_final.csv')
```
