## 2B—Explore and Clean the Data

```python
# Remove first row (index 0) because it is not an observation; it just contains units for some of the variables 
df_lithic = df_lithic.iloc[1:,]  # This takes rows 1 - end and all of the original columns 
df_lithic.head(5)
```

```python
print(df_lithic.shape) # 5299 rows and 48 columns
```

```python
print(df_lithic.size) # 5299 rows * 48 columns = 254,352 cells, wow! 
```

```python
# Investigate missingness
df_lithic.isnull().sum()
# no missing data, great! 
```

```python
# Check the unique values for each variable
df_lithic.nunique()
# Looks like hash and Filters 0-6 only have 1 unique value, rendering them pretty useless for building a ML model 
# so let's drop these columns. (I also visually looked at the values of these columns for the soil dataset 
# and they have the same value so these columns would tell us nothing)

# Let's also drop 'Krumbein Rnd' since this variable is not in our soil dataset, so we wouldn't be able to use it 
# as a feature in a ML model
```

```python
df_lithic = df_lithic.drop(labels=['hash', 'Filter0', 'Filter1', 'Filter2', 'Filter3', 'Filter4', 'Filter5', 'Filter6', 'Krumbein Rnd'], 
                           axis = 1)
print(df_lithic.shape) # notice 39 columns now whereas before there were 48 (we removed 9 columns)
```

```python
df_lithic.info() 
# uh oh! All of the variables are of type "object" (aka string) when they should be a numeric datatype
```

```python
# Let's fix the datatypes for each variable! 
df_lithic = df_lithic.apply(pd.to_numeric, errors='coerce')

df_lithic.info() 
```

```python
lithic_df_0_transparency_removed = df_lithic[df_lithic['Transparency'] != 0]
print(lithic_df_0_transparency_removed.shape)
```

```python
# Let's calculate some summary statistics!
lithic_df_0_transparency_removed.describe()
```

# 3—Soil Datasets

```python
def clean_soil_csv(df_soil):
    df_soil = df_soil.iloc[1:,]  # This takes rows 1 - end and all of the original columns because first row is not an observation
   
    # the first column and last column begin and end with quotation marks, respectively so let's fix those! 
    # First, let's fix the column names:
    df_soil = df_soil.rename(columns = {'\"Id': 'Id', 'Fiber Width\"': 'Fiber Width'})

    # Now, let's fix the cell values for the first and last columns:
    df_soil['Id'] = df_soil['Id'].str[1:]
    df_soil['Fiber Width'] = df_soil['Fiber Width'].str[:-1]
    return df_soil
```

```python
df_soil_0_125 = pd.read_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/Experiment9-SandGravel-0.125mm.csv', sep='\\t', encoding = 'iso8859_16', engine='python')
df_soil_0_25 = pd.read_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/Experiment9-SandGravel-0.25mm.csv', sep='\\t', encoding = 'iso8859_16', engine='python')
df_soil_0_5 = pd.read_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/Experiment9-SandGravel-0.5mm.csv', sep='\\t', encoding = 'iso8859_16', engine='python')
df_soil_1 = pd.read_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/Experiment9-SandGravel-1mm.csv', sep='\\t', encoding = 'iso8859_16', engine='python')
df_soil_2 = pd.read_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/Experiment9-SandGravel-2mm.csv', sep='\\t', encoding = 'iso8859_16', engine='python')
df_soil_4 = pd.read_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/Experiment9-SandGravel-4mm.csv', sep='\\t', encoding = 'iso8859_16', engine='python')
```

```python
df_soil_0_125 = clean_soil_csv(df_soil_0_125)

df_soil_0_25 = clean_soil_csv(df_soil_0_25)

df_soil_05 = clean_soil_csv(df_soil_0_5)

df_soil_1 = clean_soil_csv(df_soil_1)

df_soil_2 = clean_soil_csv(df_soil_2)

df_soil_4 = clean_soil_csv(df_soil_4)
```

```python
df_all_soil = pd.concat([df_soil_0_125, df_soil_0_25, df_soil_05, df_soil_1, df_soil_2, df_soil_4])
```

```python
print(df_all_soil.shape) # 405,040 rows and 47 columns
```

```python
print(df_all_soil.size) # 405,040 rows * 47 columns = 19,036,880 cells, wow! 
```

```python
# Investigate missingness
df_all_soil.isnull().sum()
# no missing data, great! 
```

```python
# Check the unique values for each variable
df_all_soil.nunique()
# Looks like hash and Filters 0-6 only have 1 unique value, rendering them pretty useless for building a ML model 
# so let's drop these columns. (I also visually looked at the values of these columns for the lithic dataset 
# and they have the same value so these columns would tell us nothing)
```

```python
df_all_soil = df_all_soil.drop(labels=['hash', 'Filter0', 'Filter1', 'Filter2', 'Filter3', 'Filter4', 'Filter5', 'Filter6'], 
                           axis = 1)
print(df_all_soil.shape) # notice 39 columns now whereas before there were 47 (we removed 8 columns)
```

```python
df_all_soil.info() 
# uh oh! All of the variables are of type "object" (aka string) when they should be a numeric datatype
```

```python
# Let's fix the datatypes for each variable! 
df_all_soil = df_all_soil.apply(pd.to_numeric, errors='coerce')

df_all_soil.info() 
```

```python
# Let's calculate some summary statistics!
df_all_soil.describe()
```

```python
# shuffle the rows so particles of different sizes are randomly throughout the all soil dataset instead of clumped together
df_all_soil = df_all_soil.iloc[np.random.permutation(df_all_soil.index)]
df_all_soil = df_all_soil.reset_index(drop=True)
```

```python
# Save the preprocessed soil dataset
df_all_soil.to_csv('/data/p_dsi/teams2023/team2/ancient_artifacts_data/all_soil_preprocessed.csv')
```
