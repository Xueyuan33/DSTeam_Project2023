```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

df=pd.read_csv('/Users/xueyuanli/Desktop/DS-team/tree_based_ML_features_final.csv')
df
```

```python
print(df.columns)
```

## Training data

```python
train_data = df[df['phone_generation'].isin([10, 11, 12])]
x_train= train_data.drop('claim', axis=1)
y_train = train_data['claim']
```

## Validation data

```python
val_data=df[(df['phone_generation'] == 13) & (df['days_since_release_date'] >=22) & (df['days_since_release_date'] <=190)]
x_val= val_data.drop('claim', axis=1)
y_val = val_data['claim']
```

## Testing data

```python
test_data=df[df['phone_generation']==14]
x_test= test_data.drop('claim', axis=1)
y_test = test_data['claim']
```

# XGBoost

```python
import xgboost as xgb
model=xgb.XGBRegressor(base_score=0.2, booster='gbtree', n_estimators=100, objective='reg:linear', max_depth=5, learning_rate=0.01)
model=model.fit(x_train, y_train, eval_set=[(x_train, y_train), (x_val, y_val)])

pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# compute the evaluation metrics
mse = mean_squared_error(y_test, pred)
rmse = mean_squared_error(y_test, pred, squared=False)
r2 = r2_score(y_test, pred)

print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R-squared score: {r2:.2f}")
```

# RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the classifier with the correct hyperparameters
rfc = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the classifier to the training data
rfc.fit(x_train, y_train)

# Predict the classes of the test data
y_pred = rfc.predict(x_test)

# Evaluate the performance of the classifier on the test data
from sklearn.metrics import mean_squared_error, r2_score

# compute the evaluation metrics
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R-squared score: {r2:.2f}")
```
