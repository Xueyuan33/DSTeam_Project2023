```python
# picked the holidays i starred because i feel like these are the holidays that people actually travel for to see family
newer_iphone_df['holiday'] = newer_iphone_df['weeks_monday'].apply(lambda x: 1 if x in pd.to_datetime(['2021-06-28', '2021-11-22', '2021-12-20',
                                                                             '2022-12-27', '2022-04-11', '2022-05-30',
                                                                             '2022-07-04', '2022-11-21', '2022-12-19',
                                                                             '2022-12-26']) else 0)
```

```python
release_dates = {
    'iphone x': '2017-11-03',
    'iphone 11': '2019-09-20',
    'iphone 12': '2020-10-23',
    'iphone 13': '2021-10-24',
    'iphone 14': '2022-09-16'
}
```

```python
newer_iphone_df['days_since_release_date'] = newer_iphone_df.apply(lambda x: (x['weeks_monday'] - pd.to_datetime(release_dates[x['phone_generation']])).days, axis=1)
```

```python
newer_iphone_df
```

### only care about phone_model not phone generation now and iphones 12, 13, & 14

```python
iphone_df = newer_iphone_df[(newer_iphone_df['phone_model'] == 'apple iphone 12') | (newer_iphone_df['phone_model'] == 'apple iphone 13') | (newer_iphone_df['phone_model'] == 'apple iphone 14')]
```

```python
iphone12_df = iphone_df[(iphone_df['phone_model'] == 'apple iphone 12')]
iphone12_df
```

```python
iphone12_df = iphone12_df[['days_since_release_date', 'weeks_monday', 'month', 'year', 'month_year', 'holiday','claim']]
iphone12_df 
```

```python
iphone12_df = iphone12_df.groupby(['days_since_release_date', 'weeks_monday', 'month', 'year', 'month_year', 'holiday'])['claim'].sum().reset_index()
iphone12_df = iphone12_df.sort_values(by='days_since_release_date')
iphone12_df.head(10)
```

```python
iphone12_df['avg_claim_this_month'] = iphone12_df.groupby(['year', 'month'])['claim'].transform('mean')
iphone12_df.head(10)
```

```python
from pandas.tseries.offsets import DateOffset
iphone12_shift = iphone12_df.copy(deep=True)
iphone12_shift['prev_month_year'] = iphone12_shift['month_year'] - DateOffset(months=1)
prev_month_avg_claim = iphone12_shift.groupby('prev_month_year')['claim'].mean().shift(1)
iphone12_shift = iphone12_shift.merge(prev_month_avg_claim.rename('avg_claim_last_month'), 
                            left_on='prev_month_year', right_index=True, how='left')
iphone12_shift.drop('prev_month_year', axis=1, inplace=True)
iphone12_shift.head(12)
```

```python
iphone12_shift_week = iphone12_shift.copy(deep=True)
iphone12_shift_week['1_week_ago'] = iphone12_shift_week['claim'].shift(1)
iphone12_shift_week['2_week_ago'] = iphone12_shift_week['1_week_ago'].shift(1)
iphone12_shift_week['3_week_ago'] = iphone12_shift_week['2_week_ago'].shift(1)
iphone12_shift_week['4_week_ago'] = iphone12_shift_week['3_week_ago'].shift(1)
iphone12_shift_week = iphone12_shift_week.rename(columns={'claim': 'this_week'})
iphone12_shift_week = iphone12_shift_week.dropna()
iphone12_shift_week = iphone12_shift_week.reset_index(drop=True)
iphone12_shift_week
```

```python
iphone12_shift_week.to_csv('/data/p_dsi/teams2023/team2/asurion_data/iPhone12_Final.csv', index=False, date_format = '%Y-%m-%d')
```
