```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
iphone12_df = pd.read_csv('/data/p_dsi/teams2023/team2/asurion_data/iPhone12_Final.csv', parse_dates=['weeks_monday','month_year'])
iphone13_df = pd.read_csv('/data/p_dsi/teams2023/team2/asurion_data/iPhone13_Final.csv', parse_dates=['weeks_monday','month_year'])
iphone14_df = pd.read_csv('/data/p_dsi/teams2023/team2/asurion_data/iPhone14_Final.csv', parse_dates=['weeks_monday','month_year'])
```

```python
iphone12_df = iphone12_df.drop(iphone12_df.index[-1])
iphone13_df = iphone13_df.drop(iphone13_df.index[-1])
iphone14_df = iphone14_df.drop(iphone14_df.index[-1])
```

```python
plt.plot(iphone12_df['days_since_release_date'], iphone12_df['this_week'], label='iPhone 12')
plt.plot(iphone13_df['days_since_release_date'], iphone13_df['this_week'], label='iPhone 13')
plt.plot(iphone14_df['days_since_release_date'], iphone14_df['this_week'], label='iPhone 14')

plt.xlabel('Days Since Release Date')
plt.ylabel('Claims')
plt.title('Claims Over Time for iPhone 12, 13, and 14')
plt.legend()
plt.show()
```

### The valley for iphone 13 occurs during the two holiday weeks of 2022-12-19 & 2022-12-26

```python
iphone13_less_than_150 = iphone13_df[(iphone13_df['days_since_release_date'] <= 150)]

plt.plot(iphone13_less_than_150['days_since_release_date'], iphone13_less_than_150['this_week'], label='iPhone 13')

plt.plot(iphone14_df['days_since_release_date'], iphone14_df['this_week'], label='iPhone 14')

plt.xlabel('Days Since Release Date')
plt.ylabel('Claims')
plt.title('Claims Over Time for iPhone 13 and iPhone 14')
plt.legend()
plt.show()
```
