```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
iphone_df = pd.read_csv('/data/p_dsi/teams2023/team2/asurion_data/time_series_ML_features_final.csv', parse_dates=['weeks_monday','month_year'])
```

```python
iphone10_df = iphone_df[(iphone_df['phone_model'] == 'apple iphone x')]
iphone11_df = iphone_df[(iphone_df['phone_model'] == 'apple iphone 11')]
iphone12_df = iphone_df[(iphone_df['phone_model'] == 'apple iphone 12')]
iphone13_df = iphone_df[(iphone_df['phone_model'] == 'apple iphone 13')]
iphone14_df = iphone_df[(iphone_df['phone_model'] == 'apple iphone 14')]
```

```python
iphone10_df
```

```python
plt.plot(iphone12_df['days_since_release_date'], iphone12_df['claim'], label='iPhone 12')
plt.plot(iphone13_df['days_since_release_date'], iphone13_df['claim'], label='iPhone 13')
plt.plot(iphone14_df['days_since_release_date'], iphone14_df['claim'], label='iPhone 14')

plt.xlabel('Days Since Release Date')
plt.ylabel('Claims')
plt.title('Claims Over Time for iPhone 12, 13, and 14')
plt.legend()
plt.show()
```

```python
iphone13_less_than_450 = iphone13_df[(iphone13_df['days_since_release_date'] <= 450)]
iphone13_less_than_450.tail(20)
```
