## Time Series plot for Iphone12

```python
iph12=data[data['phone model']=='apple iphone 12'] 
iph12
```

```python
new=iph12.groupby(['phone color','phone size']).count()
new
```

```python
iph12['type'] = iph12['phone color'] + ' ' + iph12['phone size']
iph=iph12[['type','claim','weeks_monday']]
iph.head(20)
```

```python
iph["type"].value_counts()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# group the data by 'type'
grouped = iph.groupby('type')


# plot the data for each group
for name, group in grouped:
    plt.plot(group['weeks_monday'],group['claim'], label=name)
    

# add a legend and labels for the x and y axis
plt.legend()
plt.xlabel('Date')
plt.ylabel('Claim')
plt.title('Iphone12')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 500)
# show the plot
plt.show()
```
